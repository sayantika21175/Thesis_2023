#Creating Drug Similarity Matrix
1. Get KEGG IDs for the drugs
2. Add KEGG IDs manually for the drugs which dont have KEGG IDs in mapping file (drug links.csv) obtained from DrugBank
3. Use the API to get SIMCOMP score from: http://rest.genome.jp/simcomp2/query1/query2/condition 
4. Convert similarity score fetched from API to matrix

```python
drugs=pd.read_csv('DataSet/drug_name_11_01_23.csv')#get the list of drugs we have
mapping_file=pd.read_csv('DataSet/drug_link.csv')# get the KEGG Compound ID, KEGG Drug Id, Drugbank ID from this
```

```python
drugs_with_KEGG_IDs = pd.merge(mapping_file, drugs, left_on='DrugBank ID',right_on='Drugbank_Id', how='right')

#drugs_with_KEGG_IDs.to_csv('drugs_withKEGGIDs.csv', index=False)
```

```python
drugs_with_KEGG_IDs=drugs_with_KEGG_IDs.reindex(columns=["DrugBank ID","Drugbank_Id","Name","KEGG Compound ID", "KEGG Drug ID"]).sort_values("DrugBank ID")
#
```
